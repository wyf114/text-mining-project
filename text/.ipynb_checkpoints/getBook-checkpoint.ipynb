```python
import os
import pandas as pd
```

```python
df = pd.read_csv('gutenberg_metadata.csv')
code = df['Code']
code
```

```python
for c in code:
    os.system('wget -m -nd -H -R "*-*" -A ".txt" http://www.gutenberg.org/'+ str(c) + '?filetypes[]=txt&langs[]=eng')
print("completed")
```
